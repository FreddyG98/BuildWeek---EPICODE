```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from unidecode import unidecode
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
url_duelune = "https://www.duelunecalzature.com/"
print(url_duelune)
```

```python
url_duelune_donna = "https://www.duelunecalzature.com/prodotti/donna/"
```

```python
items = BeautifulSoup(requests.get(url_duelune_donna).text).find_all("div", class_="item")
cat_donna = []

for item in items:
    cat_donna.append(item.text)

print(cat_donna)
```

```python
url_duelune_donna = "https://www.duelunecalzature.com/prodotti/uomo/calzature-uomo/"
```

```python
cat_uomo = ['sneakers','stivaletti','stringate','mocassini-slipon','polacchine','goodyear-welted','doppia-fibbia','sandali']
```

```python
cat_uomo
```

```python
sexx = ["donna","uomo"]
```

```python
#inizializzo
stringhe = []
prodotti = []
sesso = []
categorie = []
sotto_categorie = []
colori = []
prezzi_fin = []
prezzi_old = []
contatore = 0
```

```python
url_base = "https://www.duelunecalzature.com/prodotti/"
```

```python
url_base + sexx[0] + "/calzature-" + sexx[0] + "/" + cat_donna[0] + "-" + sexx[0] + "/?page=" + str(1)
```

```python
for sex in sexx:
    if sex == "donna":
        cat_sex = cat_donna
        num_max = 6
    else:
        cat_sex = cat_uomo
        num_max = 6
    for cat in cat_sex:
        for i in range(1, num_max):
            url_cat = url_base + sex + "/calzature-" + sex + "/" + cat + "-" + sex + "/?page=" + str(i)
            #print(url_cat)
            testi = BeautifulSoup(requests.get(url_cat).text).find_all("div", class_="product__wrapper fade-in-el")
            for testo in testi:
                #print(testo.text)
                scarpa_soup = BeautifulSoup(requests.get(testo.find("a").get("href")).text)
                titolo = scarpa_soup.find("h1", class_="uppercase roslindale normal-title black")
                titolo = titolo.text
                prodotto = titolo.split()[1]
                categoria = cat
                sotto_categoria = titolo.split()[0]
                colore = titolo.split()[-1]
                print(titolo)
                print("colore:", colore)
                print("categoria:", categoria)
                print("sotto_categoria:", sotto_categoria)
                print("prodotto:", prodotto)
                
                #prezzo finale
                prezzo_fin_long = scarpa_soup.find("span", class_="woocommerce-Price-amount amount").find("bdi").text
                if prezzo_fin_long is None:
                    prezzo_fin = np.NaN
                else:
                    prezzo_fin = float(prezzo_fin_long.split()[0].replace(",","."))
                print("prezzo finale:", prezzo_fin)
                
                #prezzo iniziale
                prezzo_old_long = scarpa_soup.find("span", class_="screen-reader-text")
                if prezzo_old_long is None:
                    prezzo_old = prezzo_fin
                else:
                    prezzo_old = float(prezzo_old_long.text[25:30].replace(",","."))
                print("prezzo iniziale:", prezzo_old)
                print("\n")

                stringhe.append(titolo)
                prodotti.append(prodotto)
                sesso.append(sex)
                categorie.append(categoria)
                sotto_categorie.append(sotto_categoria)
                colori.append(colore)    
                prezzi_fin.append(prezzo_fin)
                prezzi_old.append(prezzo_old)
                
```

```python
print(len(stringhe))
print(len(colori))
print(len(prezzi_old))   
print(len(prodotti))     #seconda parola
```

```python
duelune_df = pd.DataFrame(
    {"Info": stringhe,
     "Nome": prodotti,
     "Sesso": sesso,
     "Categoria": categorie,
     "Sotto_categoria": sotto_categorie,
     "Colore": colori,
     "Prezzo_finale": prezzi_fin,
     "Prezzo_originale": prezzi_old}
)
```

```python
duelune_df
```

```python
duelune_df["Sconto"] = round(((duelune_df.Prezzo_originale - duelune_df.Prezzo_finale)/duelune_df.Prezzo_originale)*100)
```

```python
duelune_df.loc[duelune_df.Sconto<0,"Prezzo_originale"] = duelune_df.loc[duelune_df.Sconto<0,"Prezzo_originale"].replace(duelune_df.loc[duelune_df.Sconto<0,"Prezzo_finale"])
```

```python
duelune_df.to_csv("duelune.csv", index=None)
```
